```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_csv('data/eda_data.csv', index_col=0)
```

```python
# Dropping unnecessary columns
df.drop(['id','fltid','arr_iata','dep_iata'],axis=1,inplace=True)
```

```python
categorical_columns = ['depstn','arrstn','status','ac','dep_country','arr_country','season', 'airline_code']

# Converting strings to numerical values
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
for i in categorical_columns:
    df[i]= label_encoder.fit_transform(df[i])

df.head().T
```

```python
df.to_csv('data/labelencoded_data.csv', index=True)
```
